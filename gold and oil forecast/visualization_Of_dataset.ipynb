```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
oil=pd.read_csv('oil_price.csv',parse_dates =["Date"], index_col ="Date")
oil_MA1=oil.rolling(365).mean()
oil_MA5=oil.rolling(5*365).mean()
```

```python
plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(30,10))
plt.plot(oil, color = 'red',linewidth=3, label = 'oil price distribution')
plt.plot(oil_MA1, color = 'blue', linewidth=5,label = '1 year moving average oil price')
plt.plot(oil_MA5, color = 'green', linewidth=5,label = '5 year moving average oil price')
#plt.title('Brent oil price data visualization')
plt.xlabel('DATE')
plt.ylabel('price of oil per Barrel in US Dollar')
plt.legend()
plt.show()
```
