```python
def predict_1dcnn(model,input,r_data,scaler,look_back=8,s=4):
    size=r_data.size
    predicted_1y=[]
    for i in range(size):
        #print(input)
        input_lstm=np.reshape(input[i:look_back+i],newshape=1*look_back)
        input_lstm=np.reshape(input_lstm, (1,s,2,1))
        #print(input_lstm)
        y=model.predict(input_lstm,verbose=0)
        x=scaler.transform([r_data[i]])
        input=np.append(input,x)
        y=scaler.inverse_transform(y)
        predicted_1y.append(y[0])
        #print(predicted_y)
        #i=i+1

    return predicted_1y
```

```python
def predict_1mcnn(model,input,r_data,scaler,look_back=8,s=4):
    size=r_data.size
    predicted_1y=[]
    for i in range(size):
        #print(input)
        input_lstm=np.reshape(input[i:look_back+i],newshape=1*look_back)
        input_lstm=np.reshape(input_lstm, (1,s,2,1))
        #print(input_lstm)
        y=model.predict(input_lstm,verbose=0)
        input=np.append(input,y[0])
        y=scaler.inverse_transform(y)
        predicted_1y.append(y[0])
        #print(predicted_y)
        #i=i+1

    return predicted_1y
```

```python
def predict_1m(model,input,r_data,scaler,look_back=8):
    size=r_data.size
    predicted_y=[]
    for i in range(size):
        #print(input)
        input_lstm=np.reshape(input[i:look_back+i],newshape=1*look_back)
        input_lstm=np.reshape(input_lstm, (1,look_back,1))
        #print(input_lstm)
        y=model.predict(input_lstm,verbose=0)
        #print(y)
        input=np.append(input,y[0])
        y=scaler.inverse_transform(y)
        predicted_y.append(y[0])
        #print(predicted_y)
        #i=i+1

    return predicted_y
```

```python
def predict_1d(model,input,r_data,scaler,look_back=8):
    size=r_data.size
    predicted_1y=[]
    for i in range(size):
        #print(input)
        input_lstm=np.reshape(input[i:look_back+i],newshape=1*look_back)
        input_lstm=np.reshape(input_lstm, (1,look_back,1))
        #print(input_lstm)
        y=model.predict(input_lstm,verbose=0)
        #print(y)
        x=scaler.transform([r_data[i]])
        input=np.append(input,x)
        y=scaler.inverse_transform(y)
        predicted_1y.append(y[0])
        #print(predicted_y)
        #i=i+1

    return predicted_1y
    
```

```python
def create_dataframe(x,r_data,a_price):
    l=r_data.size
    date=a_price['Date']
    date=date.tail(l)
    p=pd.DataFrame(x,index=date)
    p.columns=['price']
    return p
```

```python
def rootms(r_data,x):
    from sklearn.metrics import mean_squared_error
    from math import sqrt
    rms = sqrt(mean_squared_error(r_data, x))
    return rms
```

```python
#predicting 1 month ahead gold price
a_price=pd.read_csv("oil_price.csv",parse_dates =["Date"])
r_data=pd.read_csv('oil_july.csv').values
dataset=pd.read_csv('oil_train.csv')
data=dataset.tail(8)
data = np.asarray(data[['Price']])
scaler = MinMaxScaler(feature_range=(0, 1))
x_sample = [8, 150]
scaler.fit(np.array(x_sample)[:, np.newaxis])
input_m=scaler.transform(data)
gru_gold=pickle.load(open('Gru_oil.sav','rb'))
vl_gold=pickle.load(open('vanilla_lstm_oil.sav','rb'))
sl_gold=pickle.load(open('stack_lstm_oil.sav','rb'))
cn_gold=pickle.load(open('cnn_lstm_oil.sav','rb'))
y_gru=np.vstack(predict_1m(gru_gold,input_m,r_data,scaler))
y_vl=np.vstack(predict_1m(vl_gold,input_m,r_data,scaler))
y_sl=np.vstack(predict_1m(sl_gold,input_m,r_data,scaler))
y_cn=np.vstack(predict_1mcnn(cn_gold,input_m,r_data,scaler))
#print(y)
```

```python
#getting rmse 
print("Vanilla lstm rmse=",rootms(r_data,y_vl))
print("Stacked lstm rmse=",rootms(r_data,y_sl))
print("GRU rmse=",rootms(r_data,y_gru))
print("CNN lstm rmse=",rootms(r_data,y_cn))
```

```python
#creating plot
Y_vanilla=create_dataframe(y_vl,r_data,a_price)
Y_stacked=create_dataframe(y_sl,r_data,a_price)
Y_gru=create_dataframe(y_gru,r_data,a_price)
Y_cnn=create_dataframe(y_cn,r_data,a_price)
gold=pd.read_csv('oil_price.csv',parse_dates =["Date"], index_col ="Date")
real_data=gold.tail(r_data.size)
#print(real_data)
#print(Y_vanilla)
#print(Y_gru)
#print(Y_stacked)
#print(Y_cnn)
plt.figure(figsize=(30,8))
plt.rcParams.update({'font.size': 22})
plt.plot(real_data, color = 'black',linewidth=2, label = 'Real oil price')
plt.plot(Y_vanilla, color = 'red',linewidth=2, label = 'Predicted oil price using Vanilla lstm')
plt.plot(Y_stacked, color = 'blue',linewidth=2, label = 'Predicted oil price using stacked lstm')
plt.plot(Y_gru, color = 'green', linewidth=2,label = 'Predicted oil price using GRU')
plt.plot(Y_cnn, color = 'm', linewidth=2,label = 'Predicted oil price using CNN lstm')
#plt.title('one day ahead gold price prediction using stacked LSTM')
plt.xlabel('Date')
plt.ylabel('oil price')
plt.legend()
plt.show()
```
